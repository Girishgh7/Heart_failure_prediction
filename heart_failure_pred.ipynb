```python
#importing libs
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import kagglehub
```

```python
#loading datasets
df=pd.read_csv("/content/heart.csv")
df
```

```python
# Handle missing values (replace with mean for numerical features)
for column in df.select_dtypes(include=np.number).columns:
    df[column] = df[column].fillna(df[column].mean())

# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Display the cleaned dataset
df
```

```python
df.isnull().sum()# checking for null values
```

```python
df=pd.get_dummies(df,dtype=int)
df
```

```python
df=df.astype(int)
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
print(x[1])
print(y)
```

```python
from  sklearn.preprocessing import LabelEncoder
le1=LabelEncoder()
le2=LabelEncoder()
le6=LabelEncoder()
le8=LabelEncoder()
le10=LabelEncoder()
x[:,1] = le1.fit_transform(x[:,1])
x[:,2] = le2.fit_transform(x[:,2])
x[:,6] = le6.fit_transform(x[:,6])
x[:,8] = le8.fit_transform(x[:,8])
x[:,10] = le10.fit_transform(x[:,10])
print(x)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
print(x_train)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
print(x_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree_classifier=DecisionTreeClassifier()
tree_classifier.fit(x_train,y_train)
print(x_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
neighbors_classifier=KNeighborsClassifier()
neighbors_classifier.fit(x_train,y_train)
print(x_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
forest_classifier=RandomForestClassifier()
forest_classifier.fit(x_train,y_train)
print(x_train)
```

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(x_train,y_train)
print(x_train)
```

```python
from sklearn.svm import SVC
model_svm = SVC()
model_svm.fit(x_train,y_train)
print(x_train)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
y_pred=classifier.predict(x_test)
cm=confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```

```python
y_pred_logistic = classifier.predict(x_test)
y_pred_neighbors = neighbors_classifier.predict(x_test)
y_pred_svm = model_svm.predict(x_test)
y_pred_decision = tree_classifier.predict(x_test)
y_pred_random = forest_classifier.predict(x_test)
y_pred_svm = model_svm.predict(x_test)
```

```python
logistic_accuracy = accuracy_score(y_test, y_pred_logistic)
neighbors_accuracy = accuracy_score(y_test, y_pred_neighbors)
svm_accuracy = accuracy_score(y_test, y_pred_svm)
decision_accuracy = accuracy_score(y_test, y_pred_decision)
random_accuracy = accuracy_score(y_test, y_pred_random)
svm_accuracy = accuracy_score(y_test, y_pred_svm)
```

```python
print(f"Logstic Regression {logistic_accuracy}\n")
print(f"KNN {neighbors_accuracy}\n")
print(f"SVM {svm_accuracy}\n")
print(f"Decision Tree {decision_accuracy}\n")
print(f"Random Forest {random_accuracy}\n")
```

```python
plt.figure(figsize=(10, 9))
colors = ['red', 'blue', 'green', 'orange', 'purple']
plt.bar(['Logistic Regression', 'KNN', 'SVM', 'Decision Tree', 'Random Forest'],
        [logistic_accuracy, neighbors_accuracy, svm_accuracy, decision_accuracy, random_accuracy],color=colors,alpha=0.87)
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.title('Model Comparison')
plt.ylim(0, 1)
plt.show()
```
